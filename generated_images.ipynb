```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model('adv.h5')
```

```python
a = model(np.random.normal(size = (9, 100)))
```

```python
figure = plt.figure(figsize=(10, 8))
for row in range(5):
    for col in range(5):
        figure.add_subplot(3, 3, row+col+1)
        plt.imshow(a[row +col], interpolation="none")
```
